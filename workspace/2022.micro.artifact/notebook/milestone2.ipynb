```python
import pandas as pd
import numpy as np
import os
import re
import matplotlib.pyplot as plt

# Function to parse stats directly from the text file
def parse_stats_file(stats_file):
    if not os.path.exists(stats_file):
        print(f"Warning: Stats file not found: {stats_file}")
        return None
    
    stats = {'level_energy': {}}
    try:
        with open(stats_file, 'r') as f:
            content = f.read()
            
            # Extract total energy from Summary Stats
            summary_energy_match = re.search(r'Summary Stats\n-{3,}\n.*?Energy:\s+([\d.]+)\s+uJ', content, re.DOTALL)
            if summary_energy_match:
                stats['energy'] = float(summary_energy_match.group(1)) * 1_000_000 # Convert uJ to pJ
            else:
                # Fallback: Try finding any "Energy (total)" line
                energy_match = re.search(r'Energy \(total\)\s+:\s+([\d.]+)', content)
                if energy_match:
                    print(f"Warning: Using fallback total energy parsing for {stats_file}")
                    stats['energy'] = float(energy_match.group(1))
                else:
                    print(f"Error: Could not parse total energy from {stats_file}")
                    stats['energy'] = 0

            # Extract cycles
            summary_cycles_match = re.search(r'Summary Stats\n-{3,}\n.*?Cycles:\s+(\d+)', content, re.DOTALL)
            if summary_cycles_match:
                stats['cycles'] = int(summary_cycles_match.group(1))
            else:
                # Fallback to first Cycles match
                cycles_match = re.search(r'Cycles\s+:\s+(\d+)', content)
                if cycles_match:
                    print(f"Warning: Using fallback cycles parsing for {stats_file}")
                    stats['cycles'] = int(cycles_match.group(1))
                else:
                    stats['cycles'] = 0

            # Extract utilization
            summary_util_match = re.search(r'Summary Stats\n-{3,}\n.*?Utilization:\s+([\d.]+)', content, re.DOTALL)
            if summary_util_match:
                stats['utilization'] = float(summary_util_match.group(1))
            else:
                # Fallback: MAC level utilization
                util_match = re.search(r'Level 0\n-{3,}\n=== MAC ===.*?Utilized instances \(average\)\s+:\s+([\d.]+)', content, re.DOTALL)
                if util_match:
                    print(f"Warning: Using fallback MAC utilization for {stats_file}. Value is avg instances, not percentage.")
                    stats['utilization'] = float(util_match.group(1))
                else:
                    stats['utilization'] = 0
                
            # Extract per-level energy
            level_pattern = r'(Level\s+(\d+)\n-{3,}\n=== ([^\n]+) ===.*?)(?=(Level\s+\d+\n-{3,})|\Z)'
            level_matches = re.finditer(level_pattern, content, re.DOTALL)
            
            level_energy = {}
            parsed_sum = 0.0
            for match in level_matches:
                level_full_header = match.group(1).strip()
                level_num = int(match.group(2))
                level_name = match.group(3).strip()
                level_content = match.group(1)
                
                energy_values = re.findall(r'Energy \(total\)\s+:\s+([\d.]+) pJ', level_content)
                
                total_level_energy = sum(float(e.replace(',', '')) for e in energy_values)
                
                level_key = f"L{level_num}_{level_name}"
                if total_level_energy > 0:
                    level_energy[level_key] = total_level_energy
                    parsed_sum += total_level_energy
                elif level_full_header:
                    level_energy[level_key] = 0.0

            stats['level_energy'] = level_energy
                
        # Validate parsed energy
        if stats['energy'] > 0 and not np.isclose(parsed_sum, stats['energy'], rtol=0.1):
            print(f"Warning: Sum of parsed level energies ({parsed_sum:,.2f} pJ) doesn't closely match total summary energy ({stats['energy']:,.2f} pJ) for {stats_file}.")
        elif stats['energy'] == 0 and parsed_sum > 0:
            print(f"Warning: Total summary energy is 0, but parsed level energy sum is {parsed_sum:,.2f} pJ for {stats_file}.")

        return stats
    except Exception as e:
        print(f"Critical Error parsing {stats_file}: {e}")
        return None

# Load and process the results
workloads = ["resnet50_conv1", "alexnet_conv1", "mobilenet_conv1"]
workload_display_names = {
    'resnet50_conv1': 'ResNet-50',
    'alexnet_conv1': 'AlexNet',
    'mobilenet_conv1': 'MobileNet'
}

structured_results = {}
unstructured_results = {}

for workload in workloads:
    # Structured 2:4 sparse architecture results
    structured_stats_file = f"../evaluation_setups/unstructured_sparse_eval/outputs/structured_2_4_{workload}/timeloop-mapper.stats.txt"
    stats = parse_stats_file(structured_stats_file)
    if stats:
        structured_results[workload] = stats
    
    # Unstructured sparse architecture results
    unstructured_stats_file = f"../evaluation_setups/unstructured_sparse_eval/outputs/unstructured_{workload}/timeloop-mapper.stats.txt"
    stats = parse_stats_file(unstructured_stats_file)
    if stats:
        unstructured_results[workload] = stats

# Create DataFrames for comparison
def create_dataframe(results, workloads):
    data = []
    for workload in workloads:
        if workload in results:
            row = {
                'Workload': workload_display_names[workload],
                'Cycles': results[workload].get('cycles', 0),
                'Energy (pJ)': results[workload].get('energy', 0),
                'Utilization (%)': results[workload].get('utilization', 0) * 100
            }
            data.append(row)
    return pd.DataFrame(data)

structured_df = create_dataframe(structured_results, workloads)
unstructured_df = create_dataframe(unstructured_results, workloads)

# Calculate comparison ratios
if not structured_df.empty and not unstructured_df.empty:
    common_workloads = set(structured_df['Workload']).intersection(set(unstructured_df['Workload']))
    structured_filtered = structured_df[structured_df['Workload'].isin(common_workloads)]
    unstructured_filtered = unstructured_df[unstructured_df['Workload'].isin(common_workloads)]
    
    structured_filtered = structured_filtered.sort_values('Workload').reset_index(drop=True)
    unstructured_filtered = unstructured_filtered.sort_values('Workload').reset_index(drop=True)
    
    comparison_df = pd.DataFrame({
        'Workload': structured_filtered['Workload'],
        'Cycle Speedup (Unstr/Struct)': unstructured_filtered['Cycles'] / structured_filtered['Cycles'].replace(0, np.nan),
        'Energy Efficiency (Unstr/Struct)': unstructured_filtered['Energy (pJ)'] / structured_filtered['Energy (pJ)'].replace(0, np.nan),
        'Utilization Ratio (Struct/Unstr)': structured_filtered['Utilization (%)'] / unstructured_filtered['Utilization (%)'].replace(0, np.nan)
    })
else:
    comparison_df = pd.DataFrame(columns=['Workload', 'Cycle Speedup (Unstr/Struct)', 
                                          'Energy Efficiency (Unstr/Struct)', 'Utilization Ratio (Struct/Unstr)'])

# Energy breakdown by hardware level
print("Energy Breakdown by Hardware Level:")
print("=" * 60)

for arch_name, results in [('Structured 2:4', structured_results), ('Unstructured', unstructured_results)]:
    print(f"\n{arch_name} Architecture:")
    
    for workload in workloads:
        if workload in results and 'level_energy' in results[workload]:
            level_energy = results[workload]['level_energy']
            total_energy = results[workload].get('energy', 0)
            
            if level_energy and total_energy > 0:
                print(f"\n{workload_display_names[workload]}:")
                
                for level, energy in level_energy.items():
                    percentage = (energy / total_energy) * 100
                    print(f"  {level}: {energy:,.2f} pJ ({percentage:.2f}%)")
            else:
                print(f"\n{workload_display_names[workload]}: Energy breakdown not available")

# Display the summary analysis
print("\nPerformance Comparison: Structured 2:4 vs Unstructured")
print("=" * 60)
print("Structured 2:4 Sparse Architecture:")
if not structured_df.empty:
    print(structured_df.to_string(index=False))
else:
    print("No data available")

print("\nUnstructured Sparse Architecture:")
if not unstructured_df.empty:
    print(unstructured_df.to_string(index=False))
else:
    print("No data available")

print("\nComparison Ratios:")
if not comparison_df.empty:
    print(comparison_df.to_string(index=False))
else:
    print("No data available for comparison")

print("\nKey Observations:")
print("-" * 60)

if not comparison_df.empty:
    cycle_ratios = comparison_df['Cycle Speedup (Unstr/Struct)'].replace([np.inf, -np.inf], np.nan).dropna()
    energy_ratios = comparison_df['Energy Efficiency (Unstr/Struct)'].replace([np.inf, -np.inf], np.nan).dropna()
    util_ratios = comparison_df['Utilization Ratio (Struct/Unstr)'].replace([np.inf, -np.inf], np.nan).dropna()
    
    if not cycle_ratios.empty:
        avg_cycle_ratio = cycle_ratios.mean()
        if avg_cycle_ratio > 1:
            print(f"1. Structured 2:4 is more cycle-efficient by {avg_cycle_ratio:.2f}x compared to unstructured")
        else:
            print(f"1. Unstructured is more cycle-efficient by {1/avg_cycle_ratio:.2f}x compared to structured 2:4")
    else:
        print("1. Cannot calculate cycle efficiency comparison (insufficient data)")
    
    if not energy_ratios.empty:
        avg_energy_ratio = energy_ratios.mean()
        if avg_energy_ratio > 1:
            print(f"2. Structured 2:4 is more energy-efficient by {avg_energy_ratio:.2f}x compared to unstructured")
        else:
            print(f"2. Unstructured is more energy-efficient by {1/avg_energy_ratio:.2f}x compared to structured 2:4")
    else:
        print("2. Cannot calculate energy efficiency comparison (insufficient data)")
    
    if not util_ratios.empty:
        avg_util_ratio = util_ratios.mean()
        if avg_util_ratio > 1:
            print(f"3. Structured 2:4 has {avg_util_ratio:.2f}x better hardware utilization compared to unstructured")
        else:
            print(f"3. Unstructured has {1/avg_util_ratio:.2f}x better hardware utilization compared to structured 2:4")
    else:
        print("3. Cannot calculate utilization comparison (insufficient data)")
    
    print("\n4. Workload-Specific Impact:")
    for _, row in comparison_df.iterrows():
        cycle_ratio = row['Cycle Speedup (Unstr/Struct)']
        energy_ratio = row['Energy Efficiency (Unstr/Struct)']
        util_ratio = row['Utilization Ratio (Struct/Unstr)']
        
        result_str = f"   - {row['Workload']}: "
        result_parts = []
        
        if not pd.isna(cycle_ratio) and not np.isinf(cycle_ratio):
            if cycle_ratio > 1:
                result_parts.append(f"{cycle_ratio:.2f}x faster with structured")
            else:
                result_parts.append(f"{1/cycle_ratio:.2f}x faster with unstructured")
        
        if not pd.isna(energy_ratio) and not np.isinf(energy_ratio):
            if energy_ratio > 1:
                result_parts.append(f"{energy_ratio:.2f}x more energy efficient with structured")
            else:
                result_parts.append(f"{1/energy_ratio:.2f}x more energy efficient with unstructured")
        
        if not pd.isna(util_ratio) and not np.isinf(util_ratio):
            if util_ratio > 1:
                result_parts.append(f"{util_ratio:.2f}x better utilization with structured")
            else:
                result_parts.append(f"{1/util_ratio:.2f}x better utilization with unstructured")
        
        if result_parts:
            result_str += ", ".join(result_parts)
        else:
            result_str += "insufficient data for comparison"
            
        print(result_str)
else:
    print("Not enough data available for meaningful observations")

print("\n5. Advantages of Structured 2:4 Sparsity over Unstructured:")
print("   - More predictable memory access patterns")
print("   - Simplified hardware implementation with predetermined 2:4 sparsity patterns")
print("   - Lower metadata overhead for encoding sparse patterns")
print("   - Better balance between model compression and computational efficiency")
print("   - Improved memory throughput due to regular access patterns")
```
